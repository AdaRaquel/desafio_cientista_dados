```python
#Separando treino e validação

train = dataset_proc.loc[dataset_proc.date < '2015-09']
valid = dataset_proc.loc[dataset_proc.date >= '2015-09']

train.shape, train.columns, valid.shape
```

```python
#A função abaixo aplica preprocessamentos de acordo com o tipo do dados:
    #1- Gerar one-hot encoding para colunas categoricas
    #2- Gerar novas variaveis com base nas ja existentes
    #3- Substituir colunas numerica pelo log dos valores
    #4- Padronizar os dados numericos
    
def preproc_dataset(dataset_proc):
    dataset = dataset_proc
    categorical_columns = ['demo2', 'demo3', 'demo4', 'demo6', 'job4', 'debt1']
    numerical_columns = ['demo1', 'demo5', 'job3', 'income2', 'debt3']
    logical_columns = ['job1', 'job2', 'income1', 'debt2']


    for col in categorical_columns:
        print(col, '=============================')
        percent = dataset[col].value_counts()/dataset.shape[0]
        total_by_cat = dataset[col].value_counts()
        total_df = pd.concat([total_by_cat, percent], axis=1, keys = ['Count', 'Percent'])
        print(total_df)


    #Renomeando categoria '-1', para remover o caractere especial do nome da futura coluna one-hot
    #dataset['demo2'] = dataset['demo2'].apply(lambda value: 'neg' if value == -1 else value)
    #dataset['debt1'] = dataset['debt1'].apply(lambda value: 'neg' if value == -1 else value)

    #1- Gerar one-hot encoding para colunas categoricas
    for col in categorical_columns:    
        col_dummies = pd.get_dummies(dataset[col], prefix = col)
        dataset = dataset.drop(col, axis = 1).join(col_dummies)

    for col in logical_columns:
        print(col, '=============================')
        percent = dataset[col].value_counts()/len(dataset)

        total_by_cat = dataset[col].value_counts()
        total_df = pd.concat([total_by_cat, percent], axis=1, keys = ['Count', 'Percent'])
        print(total_df)

    #2- Gerar novas variaveis com base nas ja existentes
    #Talvez o dia em que o credito foi solicitado seja importante
    dataset['day'] = dataset.date.apply(lambda value: int(value.split('-')[-1]))
    dataset['before_10_day'] = dataset['day'].apply(lambda value: 1 if value<=10 else 0)
    dataset['before_15_day'] = dataset['day'].apply(lambda value: 1 if value<=15 else 0)
    dataset = dataset.drop(['date'], axis = 1)

    '''for col in numerical_columns:
        if col != 'debt3':
            tresh = dataset[col].describe()['75%']
            #dataset[col+'_greater_'+str(int(tresh))] = dataset[col].apply(lambda value: 1 if value>tresh else 0)
            dataset[col+'_greater_'+str('tresh')] = dataset[col].apply(lambda value: 1 if value>tresh else 0)
            #dataset.drop(dataset[dataset[col] > tresh].index, inplace=True)
        else:
            tresh = dataset[col].describe()['25%']
            #dataset.drop(dataset[dataset[col] < tresh].index, inplace=True)

            dataset[col+'_less_'+str('tresh')] = dataset[col].apply(lambda value: 1 if value<tresh else 0)'''
    #3- Substituir colunas numerica pelo log dos valores
    for i in range(len(numerical_columns)):
        col = numerical_columns[i]
        try:
            dataset.loc[dataset_proc[col] > 0, col] = np.log(dataset_proc[col]) 
            #dataset = dataset.drop(col, axis = 1)
        except:
            print('Erro: log de valores negativos da coluna', col)
            
    #4- Padronizar os dados numericos
    for col in numerical_columns:    
        dataset[col] = (dataset[col] - dataset[col].mean())/dataset[col].std()
    print('dataset.shape', dataset.shape)

    return dataset

def show_reports(model, show_ranking=True, top=10):
    top_10f = []

    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_valid)
    cm = confusion_matrix(y_valid, y_pred)

    ax= plt.subplot()
    sns.heatmap(cm, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

    # labels, title and ticks
    ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['Bons', 'Maus']); ax.yaxis.set_ticklabels(['Bons', 'Maus']);



    # Relatório de avaliação com outras métricas 
    print("Relatório de avaliação para o modelo %s:\n%s\n"
          % (model, classification_report(y_valid, y_pred)))
    
    if show_ranking:
        attrs = list(X_train.columns.values)
        importances = model.feature_importances_
        indices = np.argsort(importances)[::-1]

        # Ranking de atributos
        print("Ranking de Atributos:")
        for f in range(10):
            if importances[indices[f]] > 0:
                print("%d. %f -- %s" % (f + 1, importances[indices[f]], attrs[indices[f]]))
                top_10f.append(attrs[indices[f]])
        return top_10f
    return
```
